####  <font color= green>groupby</font>
an operation that involves some combination of splitting the object, applying a function, and combining the results. This can be used to group large amounts of data and compute operations on these groups

```python
 df.groupby(['year'])  #now it is like a special data frame 
```

```python
# getting the life average expectancy 
df.groupby(['year']) ['lifeExp'] 

#now it is series group by object
```

```python
# to group or split up every value of year pull out the life
#expectency column and then run your function
df.groupby(['year']) ['lifeExp'].mean()
```

```python
df.groupby(['year']).mean()

#here it's calculating all the mean of other fields or columns for the same year 
```

#### <font color= green>aggregation Functions:</font>
anything that takes in a bunch of values 
and Calculates and returns one value back "mean , standard deviation"

```python
import numpy as np 
df.groupby(['year']) ['pop'].agg(np.mean)
```

```python
df.groupby(['year','continent'])[['lifeExp' , 'gdpPercap']].agg(np.mean)
```

```python
#to apply indices so you are getting a regular data frame back

df.groupby(['year','continent'])[['lifeExp' , 'gdpPercap']].agg(np.mean).reset_index()
```

### 3- Variables are stored in both rows and columns

The most complicated form of messy data occurs when variables are stored in both rows and columns.

```python
wether=pd.read_csv('Desktop/scipy-2019-pandas-master/scipy-2019-pandas-master/data/weather.csv')
```

```python
wether.head()
```

- one of the symptoms that you know that you have this problem , if you end up seeing a bunch of rows where a lot of the values are repeated 
- from d1 to d31 those should be a column called day let's fix this one first


```python
weathernew=wether.melt(id_vars=['id','year' , 'month' , 'element'] , var_name ="Day" , value_name="temperature")
weathernew
```

- we want a data frame where i have columns id, year,month, temp min , tempmax 
- so i'll convert column into individual columns (opposite of melt)

```python
weathernew.pivot_table(
              
index=['id','year','month','Day'],#list of the columns that we don't want to touch 
columns='element', # column that we're going to pivot
values='temperature' #what are we going to use to fill in when we pivot out that column
).reset_index()
```
